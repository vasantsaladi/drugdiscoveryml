```python
# Import libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
from google.colab import drive
```

```python
# Assign the name of the target protein we want to gather data on to the target variable
# This will allow us to search the ChEMBL database for data regarding this protein
target = new_client.target
target_query = target.search('influenza')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
# Choose 1 target strand of the target protein for us to research on
selected_target = targets.target_chembl_id[7]
selected_target
```

```python
# Filter out the specific target protein we have choosen to a new dataframe
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.to_csv('coronavirus_01_bioactivity_data_raw.csv', index=False)
```

```python
# Create a dataframe that gets rid of columns without standard value or canoical smiles data and add it to a new dataframe
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
len(df2.canonical_smiles.unique())
```

```python
# Get rid of any duplicate columns with the same canoical smiles data
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

```python
# Filter out the data and create a new datafram with only the data that is needed
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3
```

```python
df3.to_csv('coronavirus_02_bioactivity_data_preprocessed.csv', index=False)
```

```python
df4 = pd.read_csv('coronavirus_02_bioactivity_data_preprocessed.csv')
```

```python
# We need to filter out the compounds that have a high IC50 value
bioactivity_threshold = []
for i in df4.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

```python
# Create a new dataframe with the filtered data and save it
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

```python
# Save the DataFrame as a CSV file
df5.to_csv('coronavirus_03_bioactivity_data_preprocessed.csv', index=False)
```

```python
smiles = []

for i in df.canonical_smiles.tolist():
  cpd = str(i).split('.')
  cpd_longest = max(cpd, key = len)
  smiles.append(cpd_longest)

smiles = pd.Series(smiles, name = 'canonical_smiles')
```

```python
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
# Inspired by: https://codeocean.com/explore/capsules?query=tag:data-curation

def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem)
        moldata.append(mol)

    baseData= np.arange(1,1)
    i=0
    for mol in moldata:

        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)

        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])

        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1

    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)

    return descriptors
```

```python
df_lipinski = lipinski(df_clean_smiles.canonical_smiles)
df_lipinski
```

```python
import numpy as np
import pandas as pd

def pIC50(input_df):
    pIC50 = []
    print(input_df)

    for i in input_df['standard_value_norm']:
        molar = i * (10 ** -9)  # Converts nM to M
        pIC50.append(-np.log10(molar))

    input_df['pIC50'] = pIC50
    output_df = input_df.drop('standard_value_norm', axis=1)
    return output_df
```

```python
def norm_value(input_df):
    norm = []

    for i in input_df['standard_value']:
        if i > 100000000:
            i = 100000000
        norm.append(i)

    input_df['standard_value_norm'] = norm
    # Use axis=1 to specify dropping a column
    x = input_df.drop('standard_value', axis=1)

    return x
```
